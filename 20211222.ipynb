```python
a = int(input("정수 입력 >> "))
b = int(input("정수 입력 >> "))
print(f'b:{b}') if a<b else print(f'a:{a}')
```

### 두 개의 정수를 입력받아 큰 수에서 작은 수를 뺀 결과값 출력!

```python
a = int(input("첫 번째 정수 입력 >>"))
b = int(input("두 번째 정수 입력 >>"))
print(f'두 수의 차 : {a-b}') if a>b else print(f'두 수의 차 : {b-a}')
```

### 키보드로 정수를 입력받아 홀수인지 짝수인지 판별하세요!

```python
num = int(input('정수 입력 >>'))
print(f'{num} 는(은) 홀수입니다.') if num%2 == 1 else print(f'{num} 는(은) 짝수입니다.')
```

```python
num = int(input('정수 입력>> '))
if num % 3 ==0 & num % 5 ==0 :
    print('3과 5의 배수입니다.')
else :
    print('3과 5의 배수가 아닙니다.')
```

```python
num1 = int(input('첫 번째 정수 입력 >>'))
num2 = int(input('두 번째 정수 입력 >>'))
if num1 > num2 :
    print('첫 번째 정수가 더 큽니다.')
elif num1 < num2 :
    print('두 번째 정수가 더 큽니다.')
else :
    print('두 수가 똑같습니다.')
```

```python
score = int(input('점수 입력 >> '))
if score <= 100 and score >= 90 :
    print(f'{score}점은 A학점 입니다.')
elif score <90 and score >= 80 :
    print(f'{score}점은 B학점 입니다.')
elif score <80 and score >= 70 :
    print(f'{score}점은 C학점 입니다.')
elif score <70 and score >= 60 :
    print(f'{score}점은 D학점 입니다.')
else :
    print(f'{score}점은 F학점 입니다.')
```
